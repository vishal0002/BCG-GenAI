```python
# Revenue Growth Percentage Calculation
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Revenue Growth (%)'] = df['Revenue Growth (%)'].round(2)
# Net Income Growth Percentage Calculation
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df['Net Income Growth (%)'] = df['Net Income Growth (%)'].round(2)
# Cash Flow Growth Percentage Calculation
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow'].pct_change() * 100
df['Cash Flow Growth (%)'] = df['Cash Flow Growth (%)'].round(2)
# Cash Flow Growth Calculation
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow'].pct_change() * 100
df['Cash Flow Growth (%)'] = df['Cash Flow Growth (%)'].round(2)
# YoY Liabilities Growth Calculation
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Liabilities Growth (%)'] = df['Liabilities Growth (%)'].round(2)
# Net Profit Margin Calculation
df['Net Profit Margin (%)'] = (df['Net Income'] / df['Total Revenue']) * 100
df['Net Profit Margin (%)'] = df['Net Profit Margin (%)'].round(2)
# Debt to Revenue Ratio Calculation
df['Debt to Revenue Ratio'] = df['Total Liabilities'] / df['Total Revenue']
df['Debt to Revenue Ratio'] = df['Debt to Revenue Ratio'].round(2)
# Debt to Income Ratio Calculation
df['Debt to Income Ratio'] = df['Total Liabilities'] / df['Net Income']
df['Debt to Income Ratio'] = df['Debt to Income Ratio'].round(2)
# Cash Flow to Revenue Ratio Calculation
df['Cash Flow to Revenue (%)'] = (df['Cash Flow'] / df['Total Revenue']) * 100
df['Cash Flow to Revenue (%)'] = df['Cash Flow to Revenue (%)'].round(2)
df
```

#### Outliers and Trends:
- **Outliers:**  
    - For Apple, 2023 shows a notably low revenue Z-score (-1.10), indicating an unusually weak revenue year compared to its average.
    - For Microsoft, 2024 has a high revenue Z-score (1.11), highlighting a strong revenue year.
    - For Tesla, 2022 has a low revenue Z-score (-1.15), suggesting underperformance, while 2023 and 2024 are above average.
- **Trends:**  
    - Microsoft demonstrates strong revenue growth, with YoY growth rates increasing each year and a strong upward trend in revenue Z-scores.
    - Tesla shows high volatility in both revenue growth and net income, with a large jump in revenue from 2022 to 2023, but much smaller growth in 2024.
    - Apples revenue growth is relatively flat, with a dip in 2023 and a slight recovery in 2024, resulting in an overall negative CAGR.

### Key Findings

- **Apple:** Slight decline in revenue growth over the period (CAGR -0.42%), with relatively stable profit margins and moderate fluctuations in liabilities and cash flow.
- **Microsoft:** Strong revenue growth (CAGR 11.19%), consistent improvement in revenue and cash flow, and healthy profit margins.
- **Tesla:** High revenue growth (CAGR 9.51%), but more volatility in net income and liabilities compared to peers.
